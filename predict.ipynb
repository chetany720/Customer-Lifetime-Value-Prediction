```python
import pandas as pd
```

```python
# Load the CSV file into a pandas DataFrame
df = pd.read_csv('Book1.csv')
```

```python
# Display the first few rows of the DataFrame
print(df.head())
```

```python
# Describe the DataFrame
print(df.describe())
```

```python
# Convert the 'InvoiceDate' column to datetime format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
# Extract the year, month, and day from the 'InvoiceDate' column
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
```

```python
# Group the data by 'Country' and calculate the total revenue for each country
country_revenue = df.groupby('Country')['Price'].sum().reset_index()
print(country_revenue)
```

```python
# Sort the 'country_revenue' DataFrame by the 'Price' column in descending order
country_revenue = country_revenue.sort_values(by='Price', ascending=False)
print(country_revenue)
```

```python
# Plot a bar chart of the total revenue for each country
import matplotlib.pyplot as plt
country_revenue.plot(kind='bar', x='Country', y='Price')
plt.title('Total Revenue by Country')
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.show()
```
